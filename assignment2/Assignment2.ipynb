```python
# input data file
X_train_file = dir + 'X_train.csv'
Y_train_file = dir + 'Y_train.csv'

X_test_file = dir + 'X_test.csv'
Y_test_file = dir + 'Y_test.csv'

# import data from csv file to numpy format
X_train = np.loadtxt(X_train_file)
Y_train = np.loadtxt(Y_train_file)
# reshape y matrix to match the x matrix form
Y_train = np.reshape(Y_train, (Y_train.shape[0], 1))

X_test = np.loadtxt(X_test_file)
Y_test = np.loadtxt(Y_test_file)
# reshape y matrix to match the x matrix form
Y_test = np.reshape(Y_test, (Y_test.shape[0], 1))
```

```python
# Define sigmoid activation function
def sigmoid(z):
	return 1 / (1 + np.exp(-z))
# Define derivative of sigmoid function
def sigmoid_derivative(z):
  return sigmoid(z) * (1 - sigmoid(z))
# Define tanh activation function
def tanh(z):
  return (np.exp(z) - np.exp(-z)) / (np.exp(z) + np.exp(-z))
# Define derivative of tanh function
def tanh_derivative(z):
  return 1 - np.power(tanh(z), 2)
# Define ReLU activation function
def ReLU(z):
  return np.maximum(0, z)
# Define derivative of ReLU function
def ReLU_derivative(z):
  z[z > 0] = 1
  z[z <= 0] = 0
  return z
# Define mean square error loss function
def mse_loss(y_hat, y):
  return np.square(np.subtract(y, y_hat)).mean()
```
